### Linear Reggresion & Rigde and Lasso 

```python
#house Pricing

from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=load_boston() 
```

```python
type(df)
```

```python
df
```

```python
ds=pd.DataFrame(df.data)
ds.columns=df.feature_names
ds.head()
```

```python
 ds['PRICE']=df.target
```

```python
ds
```

```python
ds.drop(['price'], axis=1,inplace=True)
```

```python
ds.drop(['ice'], axis=1,inplace=True)
```

```python
ds
```

### Dividing the dataset in dependent and independent

```python
x=ds.iloc[:,:-1]##Independent Feature
y=ds.iloc[:,-1]##Dependent Feature
```

```python
x.head()
```

```python
y.head()
```

### Linear Regression 

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(
x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
print(mse)
```

```python
np.mean(mse)
```

```python
lin_reg.fit(x,y)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ri=Ridge()
```

```python
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,5,10,20,40,60,40,80,90,100]}

rid=GridSearchCV(ri,params,scoring='neg_mean_squared_error',cv=50)
rid.fit(x,y)
```

```python
print(rid.best_params_)
print(rid.best_score_)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
li=Lasso()
```

```python
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,5,10,20,40,60,40,80,90,100,102,120,200,300]}

lid=GridSearchCV(li,params,scoring='neg_mean_squared_error',cv=50)
lid.fit(x,y)
```

```python
print(lid.best_params_)
print(lid.best_score_)
```

```python
y_pred=lid.predict(x_test)
from sklearn.metrics import r2_score

r2_score1=r2_score(y_pred,y_test)
```

```python
print(r2_score1)
```
